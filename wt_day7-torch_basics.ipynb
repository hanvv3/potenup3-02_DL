# 2. Neural Network (nn) 구현 방법

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.tensorboard import SummaryWriter



class Net(nn.Module):
    def __init__(self) -> None:
        super(Net, self).__init__()
        self.fc1 = nn.Linear(in_features=3, out_features=4, bias=True)
        self.fc1_act = nn.ReLU()
        self.fc2 = nn.Linear(in_features=4, out_features=3, bias=True)
        self.fc2_act = nn.ReLU()
        self.fc3 = nn.Linear(in_features=3, out_features=1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        
        return x
        

model = Net()
model
```

```python
import torch.nn as nn
import torch.nn.functional as F

class LRModel(nn.Module):
    def __init__(self) -> None:
        super().__init__()
        self.fc1 = nn.Linear(1,4)
        self.fc2 = nn.Linear(4,4)
        self.fc3 = nn.Linear(4,3)
        self.fc4 = nn.Linear(3,1)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        
        return x

model = LRModel()
model
```

```python
list(model.parameters())
```

```python
epochs = 500

criterion = nn.MSELoss()
optimizer = optim.Adam(
    params=model.parameters(),
    lr=1e-1
)

loss_list = []

for ep in range(epochs):
    
    optimizer.zero_grad()           # 1. 기울기 초기화
    
    y_pred = model(X)               # 2. forward propagation
    
    loss = criterion(y_pred, y)     # 3. 손실 계산
    
    loss_list.append(loss.item())   # 4. 손실 저장
    print(loss.item())              # 5. 손실 출력
    
    loss.backward()                 # 6. 역전파

    optimizer.step()                # 7. 학습률 만큼 W 업데이트
    
```
